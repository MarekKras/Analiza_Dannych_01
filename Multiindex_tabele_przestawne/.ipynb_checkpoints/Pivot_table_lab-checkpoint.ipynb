```python
import pandas as pd
import numpy as np
```

```python
from datetime import datetime
from datetime import timedelta 
import time 
```

```python
df = pd.read_csv('./Data/marathon_results_2016.csv',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],index_col='Bib')
```

```python
df.head()
```

```python
df['40K'] = df['40K'].apply(pd.to_timedelta)
```

```python
df['Half'] = df['Half'].apply(pd.to_timedelta)
```

```python
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head(1000)
```

```python
df.pivot_table(index="Age",columns="M/F",values="TotalSeconds",aggfunc='mean').head()
```

```python
kon =  df.pivot_table(index="Age",columns="M/F",values=["TotalSeconds","HalfSeconds"],aggfunc='mean').head()
```

```python
kon.head()
```

```python
kon.swaplevel(axis='columns')
```

```python
kon.swaplevel(axis='columns').sort_index(axis='columns')
```

```python
df.pivot_table(index="Age",columns="M/F",aggfunc='count',values='City').head()
```

```python
df.pivot_table(index="Age",aggfunc='count',values='City').head()
```

```python
df.pivot_table(index="Country",aggfunc='count',values='City').head()
```

```python
df.pivot_table(index="Country",aggfunc='count',values='City').sort_index(ascending=False).head()
```

```python
df.pivot_table(index=['Country','State','City'],aggfunc='mean',values='TotalSeconds').sort_values('TotalSeconds',ascending=False).head()
```
